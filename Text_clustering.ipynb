```python
import pandas as pd
import re
import nltk
import spacy
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from sentence_transformers import SentenceTransformer


# DOWNLOAD NLTK RESOURCES

nltk.download('stopwords')
stop_words = set(nltk.corpus.stopwords.words('english'))


# LOAD SPACY MODEL

nlp = spacy.load("en_core_web_sm")  # Ensure this is installed: python -m spacy download en_core_web_sm

#  CLEANING FUNCTION

def clean_text(text):
    """Removes '__eou__', URLs, mentions, punctuation, stopwords, and lemmatizes text."""
    # 1. Lowercase
    text = text.lower()
    # 2. Remove '__eou__'
    text = text.replace('__eou__', ' ')
    # 3. Remove URLs
    text = re.sub(r'http\S+|www.\S+', '', text)
    # 4. Remove mentions and numbers
    text = re.sub(r'@\w+|\d+', '', text)
    # 5. Remove punctuation/special chars
    text = re.sub(r'[^a-z\s]', '', text)

    # Tokenization
    words = text.split()
    # Remove stopwords
    words = [w for w in words if w not in stop_words]
    # Lemmatization
    words = [token.lemma_ for token in nlp(' '.join(words))]
    return ' '.join(words)


# LOAD DATA (FOR .txt FILES)

with open("dialogues_text.txt", 'r', encoding='utf-8') as f:
    lines = f.read().splitlines()

data = pd.DataFrame({'text': lines})
data['cleaned_text'] = data['text'].astype(str).apply(clean_text)



# GENERATE EMBEDDINGS

model = SentenceTransformer('all-mpnet-base-v2')
embeddings = model.encode(data['cleaned_text'].tolist(), batch_size=32, show_progress_bar=True)
data = pd.DataFrame({'text': lines})
data['cleaned_text'] = data['text'].astype(str).apply(clean_text)




```

```python


from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Fixed number of clusters
n_clusters = 8

kmeans = KMeans(n_clusters=n_clusters, random_state=42)
labels = kmeans.fit_predict(embeddings)

sil_score = silhouette_score(embeddings, labels)

# Final results
data["cluster"] = labels
print(f"✅ Number of clusters: {n_clusters}, Silhouette Score = {sil_score:.4f}")



```

```python

embedding_2d = TSNE(random_state=42).fit_transform(embeddings)

plt.figure(figsize=(12, 8))
plt.scatter(embedding_2d[:,0], embedding_2d[:,1], c=labels, cmap="Spectral", s=5)
plt.colorbar()
plt.title("t-SNE of Clusters")
plt.show()

# SAVE RESULTS

data.to_csv('clustered_data3.csv', index=False)
print("\n✅ Done! Results saved to clustered_data3.csv.")
```
